```
#importing the required liabraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
df
```

```
plt.plot([1,2,3,4],linewidth=5.0)
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```
plt.plot(df['MinTemp'],df['MaxTemp'],linewidth=5.0)
plt.ylabel('MinTemp')
plt.xlabel('MaxTemp')
plt.show()
```

```
plt.scatter(df['MinTemp'],df['MaxTemp'])
plt.ylabel('MinTemp')
plt.xlabel('MaxTemp')
plt.show()
```

```
t1=df['MinTemp']
t2=df['MaxTemp']
plt.figure(1)
plt.subplot(211)
plt.plot(t1,t1,'bo',label='t1')
plt.plot(t2,t2**2,'g',label='t2')
plt.title('subplots')
plt.legend()

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.grid(True)
plt.text(2.18,0.8,r'cos func')
plt.show()
```
